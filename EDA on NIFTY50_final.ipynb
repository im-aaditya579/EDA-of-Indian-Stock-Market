```python
df2 = pd.read_csv(r"C:\Users\abhi0\Downloads\50-stocks.csv")
```

Loading 50 stocks of NIFTY50 with their daily prices:

```python
df2.head()
```

```python
df2.info()
```

```python
df2["Date_dt"] = df2["DATE"].values.astype('datetime64[D]')
```

```python
df2["Year"] = df2["Date_dt"].dt.year
```

```python
df2["Year"].unique()
```

```python
df_new = df2.loc[df2["Year"] > 2016,:]
```

Keeping data of only the past 5 years:

```python
df_new = df_new.reset_index()
```

```python
df_new.drop("index",axis='columns',inplace=True)
```

```python
df_new.head()
```

```python
print("Shape of the dataframe is : ",df_new.shape)
```

```python
df_new.describe()
```

```python
finance_stocks = ['HDFCBANK','HDFC','SBIN','KOTAKBANK','ICICIBANK','BAJAJFINSERV','INDUSBANK','BAJAJFINANCE']
it_stocks = ['TCS','INFY','WIPRO','TECHM','HCLTECH']
energy_stocks = ['IOC','NTPC','ONGC','POWERGRID','BPCL','COALINDIA']
pharma_stocks = ['SUNPHARMA','DRREDDYS','CIPLA']
automobile_stocks = ['TATAMOTORS','MARUTI','HEROMOTOCO','EICHERMOTOR']
```

```python
df_sector = pd.DataFrame()
df_sector["NIFTY_Finance"] = df_new[finance_stocks].sum(axis=1)
df_sector["NIFTY_IT"] = df_new[it_stocks].sum(axis=1)
df_sector["NIFTY_Energy"] = df_new[energy_stocks].sum(axis=1)
df_sector["NIFTY_Pharma"] = df_new[pharma_stocks].sum(axis=1)
df_sector["NIFTY_Automobile"] = df_new[automobile_stocks].sum(axis=1)
```

```python
df_sector.head()
```

# Finance sector:

```python
plt.figure(figsize=(15,6))
sns.set_style('whitegrid')
plt.title('Finance portfolio in the last 5 years')
sns.lineplot(df_new.Year,df_new.HDFCBANK,label='HDFCBANK',color='blue').set(xlabel='Years',ylabel='stock Price');
sns.lineplot(df_new.Year,df_new.HDFC,label='HDFC',color='purple').set(xlabel='Years',ylabel='stock Price');
sns.lineplot(df_new.Year,df_new.SBIN,label="SBI",color='yellow').set(xlabel='Years',ylabel='stock Price');
sns.lineplot(df_new.Year,df_new.KOTAKBANK,label='KOTAK MAHINDRA',color='green').set(xlabel='Years',ylabel='stock Price');
sns.lineplot(df_new.Year,df_new.ICICIBANK,label='ICICI',color='red').set(xlabel='Years',ylabel='stock Price');
sns.lineplot(df_new.Year,df_new.AXISBANK,label='AXISBANK',color='gray').set(xlabel='Years',ylabel='stock Price');
sns.lineplot(df_new.Year,df_new.BAJAJFINSERV,label='BAJAJFINSERV',color='black').set(xlabel='Years',ylabel='stock Price');
sns.lineplot(df_new.Year,df_new.INDUSBANK,label='INDUSBANK',color='cyan').set(xlabel='Years',ylabel='stock Price');
sns.lineplot(df_new.Year,df_new.BAJAJFINANCE,label='BAJAJFINANCE',color='pink').set(xlabel='Years',ylabel='stock Price');
```

Conclusions: Bajaj Finserv has shown a huge growth in stock price over the last 5 years
